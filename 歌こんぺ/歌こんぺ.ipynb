```python
import pandas as pd

df=pd.read_csv('genre_labels.csv')
sample=pd.read_csv('sample_submit.csv',header=None,names=('number','genre'))
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')

```

```python
train["genre"] = train["genre"].replace(0, "country")
train["genre"] = train["genre"].replace(1, "electronic")
train["genre"] = train["genre"].replace(2, "folk")
train["genre"] = train["genre"].replace(3, "hip-hop")
train["genre"] = train["genre"].replace(4, "jazz")
train["genre"] = train["genre"].replace(5, "latin")
train["genre"] = train["genre"].replace(6, "classic")
train["genre"] = train["genre"].replace(7, "other-light-music")
train["genre"] = train["genre"].replace(8, "pop")
train["genre"] = train["genre"].replace(9, "religious")
train["genre"] = train["genre"].replace(10, "rock")

train["high-tempo"] = train["tempo"].str.split(pat="-", expand=True)[0]
train["low-tempo"] = train["tempo"].str.split(pat="-", expand=True)[1]
train = train.drop("tempo", axis=1)
train = train.drop("index", axis=1)

train
```

```python
train
```

```python
import warnings
warnings.filterwarnings("ignore")
from pycaret.classification import *

label = setup(train, target = 'genre')

compare_models()
```

```python
model1 = create_model('et')
model2 = create_model('catboost')
```

```python
tune_model1=tune_model(model1)
tune_model2=tune_model(model2)
```

```python
ensemble = blend_models([tune_model1, tune_model2])
```
